```python
import numpy as np
import pandas as pd

def numer_4(txt):
    txt = txt.replace(" ",",")
    txt = txt.replace("\n", "")
    txt = txt.replace(".","")
    
    import json
    txt = json.loads(txt)
    return txt

def numer_40(txt):
    txt = txt.replace("\n","")
    txt = txt.replace(" ",",")
    for i in range(2,10):
        txt = txt.replace(i*",","")
    txt = txt.replace(".,",",")
    
    import json
    txt = json.loads(txt)
    
    return txt

def numer_400(txt):    
    txt = txt.replace("\n","")
    txt = txt.replace(" ",",")
    for i in range(3,10):
        txt = txt.replace(i*",","")
    txt = txt[:-2] + "]"
    txt = txt.replace(".,",",")
    
    txt = json.loads(txt)
    return(txt)

def numer_4000(txt):
    txt = txt.replace(" ", ",").replace("\n","")
    for i in range(2,10):
        txt = txt.replace(i*",","")
    txt = txt.replace("0.0","0,0").replace(".0",",0").replace(".,]","]")
    txt = txt.replace(".,",",")
    
    
    import json
    txt = json.loads(txt)
    
    return(txt)

def numer_40000(t):
    t = t.replace(" ",",").replace("\n", "")

    for i in range(2,10):
        t = t.replace(i*",", ",")
    t = t.replace(".,,]","]") 
    t = t.replace(".,",",")

    import json
    t = json.loads(t)
    
    return t

df = pd.read_csv("df_env_infos.csv", index_col=[0])
for i in range(len(df["weights"])):
    try:
        df["weights"].loc[i] = numer(df["weights"].loc[i])
        print("1",i)
    except:
        try:
            df["weights"].loc[i] = numer_4(df["weights"].loc[i])
            print("2",i)
        except:
            try:
                df["weights"].loc[i] = numer_40(df["weights"].loc[i])
                print("3",i)
            except:
                try:
                    df["weights"].loc[i] = numer_400(df["weights"].loc[i])
                    print("4",i)
                except:
                    try:
                        df["weights"].loc[i] = numer_4000(df["weights"].loc[i])
                        print("5",i)
                    except:
                        try:
                            df["weights"].loc[i] = numer_40000(df["weights"].loc[i])
                            print("6",i)
                        except:
                            print(i)


```
