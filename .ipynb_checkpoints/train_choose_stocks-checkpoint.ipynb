```python
import pandas as pd
import numpy as np
```

```python
name_L = pd.read_csv("./data/15_name_train_2005_03_01_MAX.txt")["Unnamed: 0"]
```

```python
name_L
```

```python
data_df = pd.read_csv("./data/raw_data.csv")
```

```python
data_df = data_df[data_df["DATAYM"] >= "2005-03-01"]
```

```python
date = pd.DataFrame(data_df[data_df["PERFORMANCEID"] == "0P0000YSOP"]["DATAYM"].reset_index()["DATAYM"])

for i in range(len(name_L)):
    tmp = data_df[data_df["PERFORMANCEID"] == name_L[i]]["RET1M"].reset_index()["RET1M"]
    tmp = pd.DataFrame(tmp).rename(columns = {"RET1M":name_L[i]})
    
    date = pd.concat([date,tmp], axis = 1)
    
```

```python
date
```

```python
date.to_csv("./data/train_Stock_for_each_mkt.csv")
```
