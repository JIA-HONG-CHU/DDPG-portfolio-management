```python
import numpy as np
import pandas as pd

def numer_4(txt):
    txt = txt.replace(" ",",")
    txt = txt.replace("\n", "")
    txt = txt.replace(".","")
    
    import json
    txt = json.loads(txt)
    return txt

def numer_40(txt):
    txt = txt.replace("\n","")
    txt = txt.replace(" ",",")
    for i in range(2,10):
        txt = txt.replace(i*",","")
    txt = txt.replace(".,",",")
    
    import json
    txt = json.loads(txt)
    
    return txt

def numer_400(txt):    
    
    txt = txt.replace("\n","")
    txt = txt.replace(" ",",")
    for i in range(3,10):
        txt = txt.replace(i*",","")
    txt = txt[:-2] + "]"
    txt = txt.replace(".,",",")
    
    txt = json.loads(txt)
    return(txt)

def numer_4000(txt):
    txt = txt.replace(" ", ",").replace("\n","")
    for i in range(2,10):
        txt = txt.replace(i*",","")
    txt = txt.replace("0.0","0,0").replace(".0",",0").replace(".,]","]")
    txt = txt.replace(".,",",")
    
    
    import json
    txt = json.loads(txt)
    
    return(txt)

def numer_40000(t):
    t = t.replace(" ",",").replace("\n", "")

    for i in range(2,10):
        t = t.replace(i*",", ",")
    t = t.replace(".,,]","]") 
    t = t.replace(".,",",")

    import json
    t = json.loads(t)
    
    return t

def numer_400000(t):
    t = t.replace(" ",",").replace("\n","")

    for i in range(1,10):
        t = t.replace(i*",", ",")
    t = t.replace(".,,",",").replace(",]","]")
    t = t.replace(".,",",")
    import json
    t = json.loads(t)
    
    return t

def numer_4000000(t):
    t = t.replace(" ",",").replace("\n","")
    for i in range(1,10):
        t = t.replace(i*",", ",")
    t = t.replace(".,,",",").replace(",]","]")
    t = t.replace(".,",",")
    t = t.replace(",,",",")

    import json
    t =json.loads(t)
    
    return t


df = pd.read_csv("df_env_infos.csv", index_col=[0])
for i in range(len(df["weights"])):
    try:
        df["weights"].loc[i] = numer(df["weights"].loc[i])
        print("1",i)
    except:
        try:
            df["weights"].loc[i] = numer_4(df["weights"].loc[i])
            print("2",i)
        except:
            try:
                df["weights"].loc[i] = numer_40(df["weights"].loc[i])
                print("3",i)
            except:
                try:
                    df["weights"].loc[i] = numer_400(df["weights"].loc[i])
                    print("4",i)
                except:
                    try:
                        df["weights"].loc[i] = numer_4000(df["weights"].loc[i])
                        print("5",i)
                    except:
                        try:
                            df["weights"].loc[i] = numer_40000(df["weights"].loc[i])
                            print("6",i)
                        except:
                            try:
                                df["weights"].loc[i] = numer_400000(df["weights"].loc[i])
                                print("7",i)
                            except:
                                try:
                                    df["weights"].loc[i] = numer_4000000(df["weights"].loc[i])
                                    print("8",i)
                                except:
                                    print(i)
                            


```

```python
file_you_check = "Stocks_for_each_mkt.csv"
stocks = pd.read_csv(file_you_check, index_col = [0])

## preprocess
if file_you_check == "Stocks_for_each_mkt.csv":
    long = stocks.iloc[:,1:]+1
    short = -stocks.iloc[:,1:] +1

stocks = pd.concat([ stocks.iloc[:,0], long, short ], axis = 1)

M_v_weights = np.ones([1,16])/16 ## To caculate mkt value

All_weights =  df["weights"][:133]
```

```python
port_v = []
for i in range(133):
    t = [1]+list(stocks.iloc[7+i][1:].values)
    t = np.array(t)
    len(t)
    w0 = All_weights[i]
    w0 = np.array(w0)
    
    
    t = np.asarray(t, dtype='float32')
    w0 = np.asarray(w0, dtype='float32')
    
    p = np.dot(w0,t)
    port_v.append(p)

M_v = []
for i in range(133):
    t = [1]+list(stocks.iloc[7+i][1:16].values)
    t = np.array(t)
    w0 = M_v_weights
    w0 = np.array(w0)
    
    
    t = np.asarray(t, dtype='float32')
    w0 = np.asarray(w0, dtype='float32')
    
    p = np.dot(w0,t)
    M_v.append(p)
```

```python
port_v = np.cumprod(port_v)
```

```python
M_v = np.cumprod(M_v)
```

```python
#NFBM Training & Testing :Average number of new stock per month
ls2 = []
index = []
for i in All_weights:
    ls3 = []
    for j in i:
        if (j>0 and j>0.05):
            ls3.append(i.index(j))
        for k in ls3:
            if (k+15) in ls3:
                ls3.remove(k)
    ls2.append(ls3)
index.append(len(ls2[0]))
for i in range(1,len(ls2)):
    c = 0
    for k in ls2[i]:
        if k not in ls2[i-1]:
             c+=1
    index.append(c)
print(np.mean(index))
```
