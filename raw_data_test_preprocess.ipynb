```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/Monthly_Fund_Return_Test.csv", index_col = [0])
```

```python
mkt_all = pd.pivot_table(df, values = "RET1M", index = ["DATAYM"], columns = ["BNPNAME"], aggfunc = np.nanmean)
```

```python
mkt_max = pd.pivot_table(df, values = "RET1M", index = ["DATAYM"], columns = ["BNPNAME"], aggfunc = np.nanmax)
mkt_min = pd.pivot_table(df, values = "RET1M", index = ["DATAYM"], columns = ["BNPNAME"], aggfunc = np.nanmin)
```

```python
mkt_col = pd.read_csv("data/Data_preprocessoing_v4.csv", index_col = [0])
```

```python
col_L = list(mkt_col.columns) 
```

```python
print(col_L)
```

```python
pre_df = pd.DataFrame()
```

```python
mkt_all.head()
```

```python
mkt_max.head()
```

```python
mkt_min.head()
```

```python
pre_df = pd.DataFrame()
for i in range(len(col_L)):
    
    if i == 0: ##date
        pre_df[col_L[i]] = mkt_all.index
        
    if i % 5 == 1: ##open
        pre_df[col_L[i]] = np.array(mkt_all[mkt_all.columns[ (i//5) ]])
        
    if i % 5 == 2: ##high
        pre_df[col_L[i]] = np.array(mkt_max[mkt_max.columns[ (i//5) ]])
        
    if i % 5 == 3: ##low
        pre_df[col_L[i]] = np.array(mkt_min[mkt_min.columns[ (i//5) ]])
        
    if i % 5 == 4: ##close
        pre_df[col_L[i]] = np.append( np.array( mkt_all[mkt_all.columns[ (i//5) ]][1:] ), 1)
        
    if i % 5 == 0 and (i !=0) : ##volume
        pre_df[col_L[i]] = 1
```

```python
pre_df.head()
```

```python
pre_df.to_csv("0616_text_pre.csv")
```
