```python
df = pd.read_csv("./data/Data_preprocessing.csv")
df = df.drop("Unnamed: 0", axis = 1)
```

```python
L = list(df.columns)
```

```python
close_L = [4 + i*5 for i in range(15)]
```

```python
vol_L = [5 + i*5 for i in range(15)]
```

```python
Others_L = list(range(1,76))
for i in range(len(close_L)):
    Others_L.remove(close_L[i])
    Others_L.remove(vol_L[i])
```

```python
for i in Others_L:
    df.iloc[:,i] = np.cumprod(df.iloc[:,i]+1)
```
